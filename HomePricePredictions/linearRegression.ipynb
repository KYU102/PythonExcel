```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("birminghamPrices.csv")
df.head()
```

```python
%matplotlib inline 
plt.xlabel("Date")
plt.ylabel("MedianPrice")
plt.scatter(df.Date, df.MedianPrice, color="red", marker="+")
ticks = plt.xticks( df['Date'].values[::20], rotation=90)
```

```python
dateInted = df['Date'].str.replace("/","").astype(int)
print(dateInted)
```

```python
%matplotlib inline 
plt.xlabel("Date")
plt.ylabel("MedianPrice")
plt.scatter(dateInted, df.MedianPrice, color="red", marker="+")
ticks = plt.xticks(dateInted.values[::15], rotation=90)
```

```python
combined = pd.concat(dateInted, df, join = 'outer')
print(combined)
```
