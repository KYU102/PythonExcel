```python
import pandas as pd 

df = pd.read_csv('all_amazon_orders.csv')
df
```

```python
# cleaning data by replacing n/a items (NaN) with 0
df = df.fillna(0)
df.head()
```

```python
# since the total charged has a '$' the column is stored as a string. Replace $ with nothing and convert the string into a float
df['Total Charged'] = df['Total Charged'].str.replace('$', '').astype(float)
df.head()
```

```python
df['Total Charged'].sum()
```

```python
df['Total Charged'].max()
```

```python
df['Total Charged'].min()
```

```python
df['Total Charged'].median()
```

```python
# changing the date to pandas date time format
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
#creating plot bar and setting x and y. rot = x axis rotataion. figsize is width and height of graph
df.plot.bar(x='Order Date', y='Total Charged', rot=90, figsize=(60,10))
```

```python
#create a new variable dailt orders and group same date orders togetther can calculate the sum of money spent
daily_orders = df.groupby('Order Date').sum()['Total Charged']
daily_orders.head()
```

```python
daily_orders.plot.bar(figsize=(30,10))
```
